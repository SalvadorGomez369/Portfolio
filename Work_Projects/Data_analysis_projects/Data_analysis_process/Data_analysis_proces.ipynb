Create a graph from one company that gets the total kWh produced by year

```python
import pandas as pd
import matplotlib.pyplot as plt

# Collect
df = pd.read_csv('E:\\Visualstudio\\Data_analysis_projects\\Data_analysis_process\\Resources\\Solar_Energy_Production.csv')

# Convert the date column to a datetime format and assign it to a new column:
df.loc[:, 'datetime'] = pd.to_datetime(df['date'], format='%Y/%m/%d %I:%M:%S %p')

# Create a new column for the year:
df.loc[:, 'Year'] = df['datetime'].dt.year

# Use boolean indexing to extract rows with the name 'Southland Leisure Centre'
slc = df[df.loc[:,'name'] == 'Southland Leisure Centre']

# Group the DataFrame by year and calculate the sum of the kWh column for each year
df_grouped = slc.groupby('Year')['kWh'].sum().reset_index()

# Create a bar graph using matplotlib
plt.bar(df_grouped['Year'], df_grouped['kWh'])
plt.xlabel('Year')
plt.ylabel('kWh')
plt.title('Total kWh by Year from: Southland Leisure Centre')
plt.show()
```
