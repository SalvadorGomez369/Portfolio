Gets the mean of paris columns

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('E:\\Visualstudio\\Work_Projects\\Data_analysis_projects\\arbnb\\Resources\\Aemf1.csv')

# Use boolean indexing to extract rows with the name 'Southland Leisure Centre'
slc = df[df.loc[:,'City'] == 'Paris']

p = slc.mean(numeric_only=True)
# slc.median(numeric_only=True)
# slc.mode(numeric_only=True)


# create a bar graph
p.plot(kind='barh')

# add title and axis labels
plt.title('Paris mean')
plt.xlabel('Mean')
plt.ylabel('Value')

# rotate x-axis labels for better visibility
#plt.xticks(rotation=0)

# display the plot
plt.show()
```

Max and Min

```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('E:\\Visualstudio\\Work_Projects\\Data_analysis_projects\\arbnb\\Resources\\Aemf1.csv')

# Use boolean indexing to extract rows with the name 'Southland Leisure Centre'
slc = df[df.loc[:,'City'] == 'Paris']

# Gets max and min values from the city Paris
numeric_columns = slc.select_dtypes(include='number')
max_values = numeric_columns.max()
min_values = numeric_columns.min()

# max values
# print("\nMaximum values:\n", max_values)
# print("Minimum values:\n", min_values)


# create a bar graph
max_values.plot(kind='barh')

# add title and axis labels
plt.title('Paris Max values')
plt.xlabel('Max')
plt.ylabel('Value')

# rotate x-axis labels for better visibility
#plt.xticks(rotation=0)

# display the plot
plt.show()
```

```python
# min values
# print("Minimum values:\n", min_values)


# create a bar graph
min_values.plot(kind='barh')

# add title and axis labels
plt.title('Paris Min values')
plt.xlabel('Min')
plt.ylabel('Value')

# rotate x-axis labels for better visibility
#plt.xticks(rotation=0)

# display the plot
plt.show()
```

Correlation matrix graph

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('E:\\Visualstudio\\Work_Projects\\Data_analysis_projects\\arbnb\\Resources\\Aemf1.csv')

# drop the 'RowNumber' and 'CustomerId' columns
# df = df.drop(['RowNumber', 'CustomerId'], axis=1)

# create a correlation matrix
corr_matrix = df.corr(numeric_only=True)

# round the values to 2 decimal places
corr_matrix = corr_matrix.round(1)

# create a heatmap of the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

# add title
plt.title('Correlation Matrix')

# display the plot
plt.show()
```

The Attraction Index: is a measure of the overall attractiveness of a location, taking into account factors such as proximity to public transportation, schools, shopping centers, parks, and other amenities.   
The Normalised Attraction Index: is a version of this measure that has been standardized or normalized to allow for comparison across different locations.

The Restaurant Index: is a measure of the number and quality of restaurants in a particular area, which can be an important factor for renters who enjoy dining out or want convenient access to food options.   
The Normalised Restaurant Index: is a standardized version of this measure.

By considering factors such as the Attraction Index and Restaurant Index, landlords or property managers can make informed decisions about rental rates, marketing strategies, and other aspects of managing rental properties. Likewise, potential renters can use this information to help them evaluate different locations and neighborhoods to find the best fit for their needs and preferences.
