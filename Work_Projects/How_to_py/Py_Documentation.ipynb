```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, avg
import matplotlib.pyplot as plt

def load_csv_to_spark_df(file_path, column_name, column_value):
    # Iniciar la sesión de Spark
    spark = SparkSession.builder \
        .appName("Load CSV") \
        .getOrCreate()

    # Leer el archivo CSV con PySpark
    df_spark = spark.read.csv(file_path, header=True, inferSchema=True)

    # Filtrar el DataFrame solo por las filas donde 'column_name' tiene el valor 'column_value'
    return df_spark.filter(df_spark[column_name] == column_value)

def calcular_media(df):
    """
    Calcula la media de las columnas numéricas de un DataFrame.
    :param df: El DataFrame de entrada.
    :return: Un DataFrame con las medias de las columnas numéricas.
    """
    # Filtrar las columnas numéricas
    columnas_numericas = [col for col, dtype in df.dtypes if dtype in ('int', 'double', 'float')]

    # Calcular la media de cada columna numérica
    medias = df.select([avg(col).alias(col + "_media") for col in columnas_numericas])
    
    return medias

# Llamar a la función pasando la ruta del archivo CSV
file_path = "C:/Users/nya/Documents/PythonP/Portfolio-main/Work_Projects/Data_analysis_projects/arbnb/Resources/Aemf1.csv"
column_name = "City"
column_value = "Paris"
df_spark = load_csv_to_spark_df(file_path, column_name, column_value)

# Mostrar el DataFrame filtrado (opcional)
df_spark.show()

# Calcular la media de las columnas numéricas
df_medias = calcular_media(df_spark)

# Mostrar las medias
df_medias.show()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import avg
import matplotlib.pyplot as plt

def procesar_datos_y_graficar(file_path, column_name, column_value):
    # Iniciar la sesión de Spark
    spark = SparkSession.builder.appName("Load CSV").getOrCreate()
    
    # Leer el archivo CSV con PySpark y filtrar el DataFrame
    df_spark = spark.read.csv(file_path, header=True, inferSchema=True).filter(f"{column_name} == '{column_value}'")
    
    # Calcular la media de las columnas numéricas
    df_medias = df_spark.select([avg(col).alias(col + "_media") for col, dtype in df_spark.dtypes if dtype in ('int', 'double', 'float')]).collect()
    
    # Extraer los nombres de las columnas y los valores de las medias
    columnas = df_medias[0].asDict().keys()
    valores_media = [row[col] for col in columnas for row in df_medias]
    
    # Crear un gráfico de barras horizontales con matplotlib
    plt.figure(figsize=(10, 6))
    plt.barh(list(columnas), valores_media, color='skyblue')
    
    # Añadir etiquetas y título al gráfico
    plt.title(f"Media de las columnas numéricas de '{column_value}'", fontsize=16)
    plt.xlabel("Media", fontsize=14)
    plt.ylabel("Columnas", fontsize=14)
    
    # Mostrar el gráfico
    plt.tight_layout()
    plt.show()

# Llamar a la función con la ruta del archivo y los parámetros necesarios
file_path = "C:/Users/nya/Documents/PythonP/Portfolio-main/Work_Projects/Data_analysis_projects/arbnb/Resources/Aemf1.csv"
procesar_datos_y_graficar(file_path, "City", "Paris")
```
