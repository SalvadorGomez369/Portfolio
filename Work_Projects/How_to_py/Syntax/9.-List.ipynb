Access List Items

```python
# Access Items. The first item has index 0
thislist = ["apple", "banana", "cherry"]
print(thislist[1])
```

```python
# Access the last item of the list
thislist = ["apple", "banana", "cherry"]
print(thislist[-1])
```

```python
# Range of Indexes
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(thislist[2:5])
```

```python
# Check if item exists
thislist = ["apple", "banana", "cherry"]
if "banana" in thislist:
    print("Yes, 'banana' is in the fruits list") 
```

Change List Items

```python
# Change Item Value
thislist = ["apple", "banana", "cherry"]
thislist[1] = "blackcurrant" # Change the second item
print(thislist)
```

```python
# Change a range of item values
thislist = ["apple", "banana", "cherry"]
thislist[1:2] = ["blackcurrant", "watermelon"] #
print(thislist) 
```

Add List Items

```python
# Insert Items
thislist = ["apple", "banana", "cherry"]
thislist.insert(2, "watermelon")
print(thislist)
```

```python
thislist = ["apple", "banana", "cherry"]
thislist.append("orange")
print(thislist)
```

```python
# Add the elements of a list (or any iterable) to the end of the current list
my_list = [1, 2]
my_list.extend([3, 4])
print("extend:", my_list)
```

Remove List Items

```python
# Removes the item with the specified value
my_list = [1, 2, 3, 2]
my_list.remove(2)
print("remove:", my_list)
```

```python
# Removes the element at the specified position
my_list = [1, 2, 3]
my_list.pop(1) # pop() method removes the last item if index is not provided
print("pop:", my_list)
```

```python
# The del keyword also removes the specified index
thislist = ["apple", "banana", "cherry"]
del thislist[0]
print(thislist)
```

```python
# The del keyword can also delete the list completely
thislist = ["apple", "banana", "cherry"]
del thislist 
```

```python
# Removes all the elements from the list
my_list.clear()
print("clear:", my_list)
```

Sort lists

```python
# Sorts the list in ascending order 
my_list = [3, 1, 2]
my_list.sort() # Or descending order if *reverse=True* is provided
print("sort:", my_list)
```

```python
# Sorts the list ascending order alphabetically 
thislist = [ "banana", "cherry", "apple" ]
thislist.sort(reverse=True) # Or descending order if *reverse=True* is provided
print(thislist)
```

```python
# Perform a case-insensitive sort of the list
thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.sort(key = str.lower)
print(thislist)
```

```python
# Sort the list based on how close the number is to 50
def myfunc(n):
    return abs(n - 50)

thislist = [100, 50, 65, 82, 23]
thislist.sort(key = myfunc)
print(thislist)
```

```python
# Reverse the order of the list
thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.reverse()
print(thislist)
```

```python
# Make a copy of a list
thislist = ["apple", "banana", "cherry"]
mylist = thislist.copy()
print(mylist)
```

Loop Lists

```python
# loop through a list
thislist = ["apple", "banana", "cherry"]
for x in thislist:
    print(x) 
```

List Comprehension

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]

# newlist = [*expression* for *item* in *iterable* if *condition* == True]
newlist = [x for x in fruits if "a" in x]

print(newlist)
```

Join list

```python
# join two lists
list1 = [1, 2, 3]
list2 = [4, 5, 6]
result = list1 + list2
print(result)
```

```python
# join two large lists 
import itertools

list1 = [i for i in range(1, 1000001)]  # List of numbers from 1 to 1,000,000
list2 = [i for i in range(1000001, 2000001)]  # List of numbers from 1,000,001 to 2,000,000

# Using itertools.chain to efficiently combine the lists
result = itertools.chain(list1, list2)

# Convert to a list if needed
combined_list = list(result)

print(combined_list[:10])
```

```python
# Returns the index of the first element with the specified value
fruits = ['apple', 'cherry', 'banana', 'cherry']

x = fruits.index("cherry") 

print(x)
```

```python
# Returns the number of elements with the specified value
points = [1, 4, 2, 9, 7, 8, 9, 3, 1]

x = points.count(9)

print(x)
```
