```python
# Boolean values
is_sunny = True
is_raining = False
```

```python
print(bool("Hello")) # Always True except for empty objects
print(bool()) # Empty object: *None, False, 0, "", [], {}, ()*
```

```python
# Boolean expressions
print(5 > 3)  # True
print(5 == 3)  # False
print(5 < 3)  # False
```

```python
# Using booleans in conditional statements
if is_sunny:
    print("It's a sunny day!")
else:
    print("It's not sunny today.")

if is_raining:
    print("It's raining.")
else:
    print("It's not raining.")
```
