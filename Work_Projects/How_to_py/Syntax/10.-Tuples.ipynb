```python
# tuples are the same as lists but they are immutable
mytuple = ("apple", "banana", "cherry")
print(mytuple) 
```

```python
# Just convert the mytuple to a list to work with it and then convert it back to a tuple
my_tuple = ("apple", "banana", "cherry")

my_list = list(my_tuple) # Convert the tuple to a list

my_list[1] = "kiwi" # Change the second item in the list

my_tuple = tuple(my_list) # Convert the list back to a tuple

print(my_tuple) # Print the tuple
```
