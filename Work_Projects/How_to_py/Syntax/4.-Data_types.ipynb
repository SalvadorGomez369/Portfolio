```python
# Data Types

x = 3 # Data type is set when you assign a value to a variable (int)

print(x, type(x)) # type() returns the data type of the variable x
print(x, str) # Prints the variable and specify the data type as a string
```

```python
# Built-in Data Types

# Text
x = "Hello World" # str  

# Numeric
x = 20   # int 	
x = 20.5 # float 	
x = 1j   # complex 

# Sequence
x = ["apple", "banana", "cherry"] # list can be changed 	
x = ("apple", "banana", "cherry") # tuple can't be changed	 
x = range(6) # Range is a sequence of numbers starting from 0 by default and increments by 1 (by default), and stops before a specified number.

# Mapping
x = {"key" : "value", "name" : "John", "age" : 36} # dict 

# Set
x = {"apple", "banana", "cherry"} # set 	
x = frozenset({"apple", "banana", "cherry"}) # frozenset 	

# Boolean
x = True # bool 

# Binary
x = b"Hello" # bytes 	
x = bytearray(5) # bytearray 
x = memoryview(bytes(5)) # memoryview 	

# None
x = None # NoneType
```

```python
# Casting, specify a variable type

x = str("Hello World") 	 	
x = int(20) 	 	
x = float(20.5) 	 	
x = complex(1j) 		
x = list(("apple", "banana", "cherry")) 		
x = tuple(("apple", "banana", "cherry")) 		
x = range(6) 	 	
x = dict(name="John", age=36) 	 	
x = set(("apple", "banana", "cherry")) 	 	
x = frozenset(("apple", "banana", "cherry")) 		
x = bool(5)
x = bytes(5)
x = bytearray(5)	
x = memoryview(bytes(5)) 
```

```python
#Range(start at "0 default", stop before "6 Required", step "1 default")
for i in range(6):
    print(i) 
```

```python
# Range(start at 3, stop before 6, step 1)
for i in range(3, 6):
    print(i)
```

```python
# Range(start at 0, stop before 12, step 3)
for i in range(0, 12, 3):
    print(i)   
```

```python
# Using range and converting to list
my_list = list(range(0, 12, 3))
print(my_list)
```
