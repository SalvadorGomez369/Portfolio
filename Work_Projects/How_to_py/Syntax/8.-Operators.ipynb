Python Bitwise Operators

```python
import pandas as pd

x = 5  # binary: 0101
y = 3  # binary: 0011

data = {
    "Operator": ["&", "|", "^", "~", "<<", ">>"],
    "Name": ["AND", "OR", "XOR", "NOT", "Zero fill left shift", "Signed right shift"],
    "Description": [
        "Sets each bit to 1 if both bits are 1",
        "Sets each bit to 1 if one of two bits is 1",
        "Sets each bit to 1 if only one of two bits is 1",
        "Inverts all the bits 0 to 1 and every 1 to 0",
        "Shift left by pushing zeros in from the right and let the leftmost bits fall off",
        "Shift right by pushing copies of the leftmost bit in from the left, and let the rightmost bits fall off"
    ],
    "Example": ["0101 & 0011", "0101 | 0011", "0101 ^ 0011", "~0101", "0101 << 2", "0101 >> 2"],
    "Output": [
        format(x & y, '04b'),  # 0001
        format(x | y, '04b'),  # 0111
        format(x ^ y, '04b'),  # 0110
        format(~x & 0b1111, '04b'),  # 1010 (inverted and truncated to 4 bits)
        format(x << 2, '05b'),  # 10100
        format(x >> 2, '04b')  # 0001
    ]
}

df = pd.DataFrame(data)

df
```
