```python
#SYNTAX
#PEP (Python Enhancement Proposal) 8 
#Recommends using "4 SPACES" per indentation level. 

def myfunc(a, b):
    # ← IDENTATION 4 spaces
    return a + b 

print(myfunc(3, 5))
```

```python
#SYNTAX

def myfunc(a, b):
    # ↓ 2 Lines IDENTATION
    c = a + b
    return c

print(myfunc(3, 5))
```

```python
#SYNTAX

def outer_function(a, b):
    # This is the outer function, indentation level 0
    def inner_function(x, y):
        # This is the inner function, indentation level 1 (4 spaces)
        return x + y

    # Call the inner function within the outer function
    result = inner_function(a, b)
    return result

# Call the outer function
print(outer_function(3, 5))  # Output will be 8
```
