```python
# Strings are sequences of characters, using the syntax of either single quotes or double quotes
print("Hello")
print('Hello')
```

```python
# You can use double quotes inside single quotes and vice versa
print("It's alright")
print("He is called 'Johnny'")
print('He is called "Johnny"')
```

```python
# You can use triple quotes for multi-line strings
a = """Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua."""
print(a) 
```

```python
# Or three single quotes
a = '''Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua.'''
print(a) 
```

```python
# Strings are arrays of bytes representing unicode characters 
a = "Hello, World!"
print(a[1]) # Starts at 0
```

```python
# Loop through the letters in the word "banana"
for x in "banana": # Loop str
    print(x)
```

```python
# Get the length of a string
a = "Hello, World!"
print(len(a)) # len() returns the length of a string
```

```python
# Check if "difficult" is present in the following text
txt = "The best things in life are difficult!"
print("difficult" in txt) # Bool a specific str check
```

```python
# Check if "difficult" is present in the following text using an if statement

txt = "The best things in life are difficult!"
if "difficult" in txt: 
    print("Yes, 'difficult' is present.")
```

```python
# Check if "difficult" is NOT present in the following text
txt = "The best things in life are difficult!"
print("expensive" not in txt) # Bool a specific str check
```

```python
# Check if "expensive" is NOT present in the following text
txt = "The best things in life are difficult!"
if "expensive" not in txt:
    print("No, 'expensive' is NOT present.")
```

```python
# Get the characters from position 1 to position 9 (not included)
# H = 0, e = 1, l = 2, l = 3, o = 4, , = 5, space = 6, W = 7, o = 8, r = 9, l = 10, d = 11, ! = 12
b = "Hello, World!" 
print(b[1:9]) #  Slicing a str
```

```python
b = "Hello, World!"
print(b[:6]) # Slicing first to 5
```

```python
b = "Hello, World!"
print(b[2:]) # Slicing 2 to last
```

```python
# H = -13, e = -12, l = -11, l = -10, o = -9, , = -8, space = -7, W = -6, o = -5, r = -4, l = -3, d = -2, ! = -1
b = "Hello, World!" # From: "o" in "World!" (position -5)
print(b[-8:-1]) # To, but not included: "d" in "World!" (position -2):
```

```python
# Modify a string
a = "Hello, World!"
print(a.upper()) # Upper case
```

```python
a = "Hello, World!"
print(a.lower()) # Lower case 
```

```python
# Returns no space at the beginning or at the end 
a = " Hello, World! " 
print(a.strip()) # Returns "Hello, World!" 
```

```python
# Replaces a string with another string
a = "Hello, World!"
print(a.replace("H", "J"))
```

```python
# Splits the string into substrings if it finds , separator
a = "Hello, World!"
print(a.split(",")) # returns ['Hello', ' World!'] 
```

```python
# Concatenate two strings
a = "Hello"
b = "World"
c = a + b
print(c)
```

```python
# Add a space between two strings 
a = "Hello"
b = "World"
c = a + " " + b
print(c)
```

```python
# Can't combine strings and numbers
age = 25
txt = "My name is Salvador, I am " + age
print(txt)
```

```python
# F-Strings
# "f" in front of the string
# "{}" ad placeholders for variables and other operations
age = 25
txt = f"My name is Salvador, I am {age}." 
print(txt)
```

```python
# f-strings can be used to format numbers
number = 3.14159
format = f"{number:.2f}"
print(format)
```

```python
# Escape characters
txt = "The sign says: \"Welcome to the unknown!\""
print(txt)
```

```python
# Add a single quote to a string
txt = 'It\'s alright.'
print(txt)
```

```python
txt = "This will insert one \\ (backslash)."
print(txt) 
```

```python
# Output to multiple lines
txt = "Hello\nWorld!"
print(txt) 
```

```python
# \r is a carriage return. It moves the cursor to the beginning of the current line.
# This allows overwriting characters printed earlier in the same line.

txt = "012\r3\r45"
print(txt)

# First \r output = "012" \r overwrite "0"12 for "3"12 
# Second \r output = "312" \r overwrite "31"2 for "45"2
# Third output = "452"
```

```python
# \t adds horizontal space between text
txt = "Hello\tWorld!"
print(txt)
```

```python
#This example erases one character (backspace)
txt = "Hello \bWorld!"
print(txt) 
```

```python
# Backslash followed by three integers will result in a octal value
txt = "\110\145\154\154\157"
print(txt)
```

```python
# Backslash followed by an 'x' and a hex number represents a hex value
txt = "\x48\x65\x6c\x6c\x6f"
print(txt) 
```
