```python
# Global Variables

x = "awesome" # Global var

def myfunc():
    x = "fantastic" # Local var
    print("Python is " + x)

myfunc()

print("Python is " + x) 
```

```python
# Global Variables

j = "break" # Global var

def myfunc(a, b):
    global j
    j = "class" # Local var, add "global" keyword to make it global
    return a + b + j

print(myfunc("continue ","def "))
```
