Insert the data frame to a db

```python
import pandas as pd
import sqlite3

# Merged DataFrame
merged_df

# Create a connection to the SQLite database
conn = sqlite3.connect('items.db')

# Create a new table in the database
conn.execute('''CREATE TABLE items (
                id INTEGER PRIMARY KEY,
                name TEXT NOT NULL,
                grade INTEGER NOT NULL,
                icon TEXT NOT NULL,
                currentStock INTEGER NOT NULL,
                totalTrades INTEGER NOT NULL,
                basePrice INTEGER NOT NULL
              );''')

# Write the DataFrame to a new table in the database
merged_df.to_sql('items', conn, if_exists='replace')

# Close the connection
conn.close()
```

Request the variants of the name enchantment

```python
import requests
import pandas as pd

ids_to_extract = [13210,13209]

# create an empty list to store the extracted dataframes
extracted_data = []

for id in ids_to_extract:
    # substitute the ID into the URL using string formatting
    url = f"https://api.arsha.io/v2/na/GetWorldMarketSubList?id={id}&lang=en"
    
    response = requests.get(url)
    data = response.json()[:12]  # extract the relevant data from the response
    
    df = pd.DataFrame(data)
    extracted_data.append(df)  # append the dataframe to the list of extracted dataframes
    
# concatenate the list of dataframes into a single dataframe
combined_data = pd.concat(extracted_data)

combined_data
```
